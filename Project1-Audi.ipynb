```python
stats = ['Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']

new_max = 100
new_min = 0
custom_range = new_max - new_min

for stat in stats:
    max_value = df3[stat].max()
    min_value = df3[stat].min()
    value_range = max_value - min_value
    df3[stat] = df3[stat].apply(lambda x: ((x-min_value)*custom_range)/value_range + new_min)
    
df3['Pokemon'] = df3.apply(lambda row: '{}'.format(row.Name), axis=1)
    
df5 = df3.loc[:,['Pokemon','Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']]

df5.set_index('Pokemon', inplace=True)

df5.head()


labels = ['Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']
values = df5.loc['TyranitarMega Tyranitar'].tolist()

variable_total = len(labels)

angles = np.linspace(0, 2*np.pi, variable_total, endpoint=False).tolist()

values += values[:1]
angles += angles[:1]

plt.figure(figsize=(6,6))
ax = plt.subplot(polar=True)
ax.plot(angles, values, color='Blue')
ax.fill(angles, values, color='Blue', alpha=0.4)
ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles), labels)
ax.set_ylim(0, 100)
ax.set_title('Tyranitar attributes', y = 1.1)
```

```python
stats = ['Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']

new_max = 100
new_min = 0
custom_range = new_max - new_min

for stat in stats:
    max_value = df3[stat].max()
    min_value = df3[stat].min()
    value_range = max_value - min_value
    df3[stat] = df3[stat].apply(lambda x: ((x-min_value)*custom_range)/value_range + new_min)
    
df3['Pokemon'] = df3.apply(lambda row: '{}'.format(row.Name), axis=1)
    
df5 = df3.loc[:,['Pokemon','Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']]

df5.set_index('Pokemon', inplace=True)

labels = ['Speed','Attack','Defense','HP','Sp..Atk','Sp..Def']
variable_total = len(labels)

angles = np.linspace(0, 2*np.pi, variable_total, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(6,6))
ax = plt.subplot(polar=True)

def add_pokemon(Pokemon, color):
    values = df5.loc[Pokemon].tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, label=Pokemon)
    ax.fill(angles, values, color=color, alpha=0.25)

add_pokemon('CharizardMega Charizard X', 'Blue')
add_pokemon('AlakazamMega Alakazam', 'Red')
add_pokemon('BeedrillMega Beedrill', 'Green')


ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles), labels)
ax.set_ylim(0, 100)
ax.set_title('Attributes', y = 1.1)
ax.legend(loc=0, bbox_to_anchor=(1.9,1.2))
```
